### |Implementation of Exponentially Moving Average (EMA) and Downside Risk for portfolio optimization

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns

# Loading data
stock_prices_df = pd.read_csv("faang_stocks.csv",parse_dates = ['Date'], index_col="Date")

# Changing the index to a datetime type allows for easier filtering and plotting.

stock_prices_df.info()
```

```python
# Plotting the stock prices
stock_prices_df.plot(title="FAANG stock prices from years 2020-2023");
plt.show()
```

```python
daily_return = stock_prices_df.pct_change().dropna() # daily returns
```

```python
# Weight optimization for minimum volatility

# Calculate the annualized expected returns and the covariance matrix
mu = daily_return.mean() * 252
Sigma = daily_return.cov() * 252


# Efficient frontier
ef = EfficientFrontier(mu, Sigma)

# Obtain optimal weights
mv_raw_weights = ef.min_volatility()
mv_portfolio = pd.Series(ef.clean_weights())

# Minimum volatility portfolio performance
mv_portfolio_performance = ef.portfolio_performance(verbose = True, risk_free_rate = 0)
mv_portfolio_vol = mv_portfolio_performance[1]

print(mv_portfolio)
```

```python
# Weight optimization for maximum Sharpe ratio
ef = EfficientFrontier(mu, Sigma)

# Obtain optimal weights
ms_raw_weights = ef.max_sharpe(risk_free_rate = 0)
ms_portfolio = pd.Series(ms_raw_weights)

# Maximum Sharpe ratio performance
ms_portfolio_performance = ef.portfolio_performance(verbose = True, risk_free_rate = 0)
ms_portfolio_sharpe = ms_portfolio_performance[2]

print(ms_portfolio)
```

## Portfolio Optimization - Exponentially Moving Average (EMA)

```python
# Mimimum volatility

mu_ema = expected_returns.ema_historical_return(stock_prices_df, span = 252, frequency = 252)

#Exponentially Weighted Covariance (EWC)
Sigma_ewc = risk_models.exp_cov(stock_prices_df, span = 252, frequency=252)

ef_ew = EfficientFrontier(mu_ema, Sigma_ewc)

# Obtain optimal weights
ew_mv_raw_weights = ef_ew.min_volatility()
ew_mv_portfolio = pd.Series(ef_ew.clean_weights())

# Minimum volatility portfolio performance
ew_mv_portfolio_performance = ef_ew.portfolio_performance(verbose = True, risk_free_rate = 0)
ew_mv_portfolio_vol = ew_mv_portfolio_performance[1]

print(ew_mv_portfolio)
```

```python
# Maximum Sharpe ratio

ef_ew = EfficientFrontier(mu_ema, Sigma_ewc)

# Obtain optimal weights
ew_ms_raw_weights = ef_ew.max_sharpe(risk_free_rate = 0)
ew_ms_portfolio = pd.Series(ew_ms_raw_weights)

# Maximum Sharpe ratio performance
ew_ms_portfolio_performance = ef_ew.portfolio_performance(verbose = True, risk_free_rate = 0)
ew_ms_portfolio_sharpe = ew_ms_portfolio_performance[2]

print(ew_ms_portfolio)
```

## Portfolio Optimization - Downside risk for optimization (Semicovariance)

```python
# Minimum volatility - Semicovariance matrix optimization
Sigma_semi = risk_models.semicovariance(stock_prices_df, benchmark = 0, frequency = 252)

# Definitions of of target retuens with different approaches
mu = daily_return.mean() * 252
mu_ema = expected_returns.ema_historical_return(stock_prices_df, span = 252, frequency = 252)

# Different efficient frontiers
ef_semi = EfficientFrontier(mu, Sigma_semi)
ef_ew_semi = EfficientFrontier(mu_ema, Sigma_semi)

# Approach 1: Annualized expected mean returns

# Obtain optimal weights
semi_mv_raw_weights = ef_semi.min_volatility()
semi_mv_portfolio = pd.Series(semi_mv_raw_weights)

print('********************** Approach 1: Semicovariance + Annual Mean **********************')

# Minimum volatility portfolio performance
semi_mv_portfolio_performance = ef_semi.portfolio_performance(verbose = True, risk_free_rate = 0)
semi_mv_portfolio_vol = semi_mv_portfolio_performance[1]

print(semi_mv_portfolio)



# Approach 2: EMA

# Obtain optimal weights
semi_ew_mv_raw_weights = ef_ew_semi.min_volatility()
semi_ew_mv_portfolio = pd.Series(ef_ew_semi.clean_weights())

print('********************** Approach 2: Semicovariance + EMA **********************')

# Minimum volatility portfolio performance
semi_ew_mv_portfolio_performance = ef_ew_semi.portfolio_performance(verbose = True, risk_free_rate = 0)
semi_ew_mv_portfolio_vol = semi_ew_mv_portfolio_performance[1]

print(semi_ew_mv_portfolio)
```

```python
# Maximum Sharpe Ratio - Semicovariance matrix optimization
Sigma_semi = risk_models.semicovariance(stock_prices_df, benchmark = 0, frequency = 252)

# Definitions of of target retuens with different approaches
mu = daily_return.mean() * 252
mu_ema = expected_returns.ema_historical_return(stock_prices_df, span = 252, frequency = 252)

# Different efficient frontiers
ef_semi = EfficientFrontier(mu, Sigma_semi)
ef_ew_semi = EfficientFrontier(mu_ema, Sigma_semi)

# Approach 1: Annualized expected mean returns

# Obtain optimal weights
semi_ms_raw_weights = ef_semi.max_sharpe(risk_free_rate=0)
semi_ms_portfolio = pd.Series(semi_ms_raw_weights)

print('********************** Approach 1: Semicovariance + Annual Mean **********************')

# Minimum volatility portfolio performance
semi_ms_portfolio_performance = ef_semi.portfolio_performance(verbose = True, risk_free_rate = 0)
semi_ms_portfolio_vol = semi_ms_portfolio_performance[1]

print(semi_mv_portfolio)



# Approach 2: EMA

# Obtain optimal weights
semi_ew_ms_raw_weights = ef_ew_semi.max_sharpe(risk_free_rate = 0)
semi_ew_ms_portfolio = pd.Series(ef_ew_semi.clean_weights())

print('********************** Approach 2: Semicovariance + EMA **********************')

# Minimum volatility portfolio performance
semi_ew_ms_portfolio_performance = ef_ew_semi.portfolio_performance(verbose = True, risk_free_rate = 0)
semi_ew_ms_portfolio_vol = semi_ew_ms_portfolio_performance[1]

print(semi_ew_ms_portfolio)
```
