```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
# Loading data
vuaa = pd.read_csv('VUAA.DE.csv', usecols=['Date', 'Close'], parse_dates = ['Date'], index_col = 'Date')
meud = pd.read_excel('Amundi Stoxx Europe 600 UCITS ETF Acc.xlsx', 
                             usecols = ['Datum', 'Offizieller Nettoinventarwert (NAV)'],
                            parse_dates = ['Datum'], index_col = 'Datum')

# Column name change prior concatenation
vuaa.rename(columns = {'Close' : 'VUAA'}, inplace = True)
meud.rename(columns={'Offizieller Nettoinventarwert (NAV)' : 'MEUD'}, inplace = True)

etfs_hist_returns = pd.concat([vuaa, meud], axis = 1)

# Discarding missing dates
etfs_hist_returns = etfs_hist_returns.dropna(axis = 0)
etfs_hist_returns.info()
```

```python
# Plot equally-dated daily returns
etfs_hist_returns.plot()
plt.show()
```

## Variables Initialization

```python
ben_weights = np.array([0.75, 0.25])
rfr = 0.026
months = etfs_hist_returns.index.to_period('M').nunique() # more accurate time-period representation
target_return_daily = (1 + rfr)**(1/365) - 1 # conversion of annual rfr to daily interest

# Portfolio daily returns
daily_return = etfs_hist_returns.pct_change().dropna()
```

## 1. Asset Allocation Optimization  - Minimum Volatility

```python
mu = expected_returns.mean_historical_return(etfs_hist_returns)
mu_2 = daily_return.mean()*252

Sigma = risk_models.sample_cov(etfs_hist_returns)
# Sigma_2 = daily_return.cov() * 252 (same result as Sigma)

# Efficient frontier
ef = EfficientFrontier(mu, Sigma)
ef_2 = EfficientFrontier(mu_2, Sigma)

# Obtain optimal weights
mv_raw_weights = ef_2.min_volatility()
mv_portfolio = pd.Series(mv_raw_weights)

# Minimum volatility portfolio performance
mv_portfolio_performance = ef_2.portfolio_performance(verbose = True, risk_free_rate = rfr)


print(mv_portfolio)
```

## 2. Asset Allocation Optimization  - Maximum Sharpe Ratio

```python
# Weight optimization for maximum Sharpe ratio
ef = EfficientFrontier(mu, Sigma)

# Obtain optimal weights
ms_raw_weights = ef.max_sharpe(risk_free_rate = rfr)
ms_portfolio = pd.Series(ms_raw_weights)

# Maximum Sharpe ratio performance
ms_portfolio_performance = ef.portfolio_performance(verbose = True, risk_free_rate = rfr)
ms_portfolio_sharpe = ms_portfolio_performance[2]

print(ms_portfolio)
```

## 3. Asset Allocation Optimization - Minimum Volatility + Exponentially Moving Average

```python
mu_ema = expected_returns.ema_historical_return(etfs_hist_returns, span = 252, frequency = 252)

# Exponentially Weighted Covariance (EWC)
Sigma_ema = risk_models.exp_cov(etfs_hist_returns, span = 252, frequency=252)

ef_ema = EfficientFrontier(mu_ema, Sigma_ema)

# Obtain optimal weights
ema_mv_raw_weights = ef_ema.min_volatility()
ema_mv_portfolio = pd.Series(ema_mv_raw_weights)

# Minimum volatility portfolio performance
ema_mv_portfolio_performance = ef_ema.portfolio_performance(verbose = True, risk_free_rate = rfr)

print(ema_mv_portfolio)
```

## 4. Asset Allocation Optimization - Maximum Sharpe + Exponentially Moving Average

```python
ef_ema = EfficientFrontier(mu_ema, Sigma_ema)

# Obtain optimal weights
ema_ms_raw_weights = ef_ema.max_sharpe(risk_free_rate = rfr)
ema_ms_portfolio = pd.Series(ema_ms_raw_weights)

# Maximum Sharpe ratio performance
ema_ms_portfolio_performance = ef_ema.portfolio_performance(verbose = True, risk_free_rate = rfr)

print(ema_ms_portfolio)
```

## 5. Asset Allocation Optimization - Minimum Volatility + Downside risk (Semicovariance)


```python
Sigma_semi = risk_models.semicovariance(etfs_hist_returns, benchmark = target_return_daily, frequency = 252)

# Different efficient frontiers
ef_semi = EfficientFrontier(mu, Sigma_semi)
ef_ema_semi = EfficientFrontier(mu_ema, Sigma_semi)

# Approach 1: Annualized expected mean returns

# Obtain optimal weights
semi_mv_raw_weights = ef_semi.min_volatility()
semi_mv_portfolio = pd.Series(semi_mv_raw_weights)

print('********************** Approach 1: Semicovariance + Annualized Mean **********************')

# Minimum volatility portfolio performance
semi_mv_portfolio_performance = ef_semi.portfolio_performance(verbose = True, risk_free_rate = 0)
semi_mv_portfolio_vol = semi_mv_portfolio_performance[1]

print(semi_mv_portfolio)


# Approach 2: EMA

# Obtain optimal weights
semi_ema_mv_raw_weights = ef_ema_semi.min_volatility()
semi_ema_mv_portfolio = pd.Series(semi_ema_mv_raw_weights)

print('********************** Approach 2: Semicovariance + EMA **********************')

# Minimum volatility portfolio performance
semi_ema_mv_portfolio_performance = ef_ema_semi.portfolio_performance(verbose = True, risk_free_rate = rfr)

print(semi_ema_mv_portfolio)
```

## 6. Asset Allocation Optimization - Maximum Sharpe + Downside Risk (Semicovariance)

```python
Sigma_semi = risk_models.semicovariance(etfs_hist_returns, benchmark = target_return_daily, frequency = 252)

# Different efficient frontiers
ef_semi = EfficientFrontier(mu, Sigma_semi)
ef_ema_semi = EfficientFrontier(mu_ema, Sigma_semi)

# Approach 1: Annualized expected mean returns

# Obtain optimal weights
semi_ms_raw_weights = ef_semi.max_sharpe(risk_free_rate=rfr)
semi_ms_portfolio = pd.Series(semi_ms_raw_weights)

print('********************** Approach 1: Semicovariance + Annualized Mean **********************')

# Minimum volatility portfolio performance
semi_ms_portfolio_performance = ef_semi.portfolio_performance(verbose = True, risk_free_rate = rfr)

print(semi_mv_portfolio)



# Approach 2: EMA

# Obtain optimal weights
semi_ema_ms_raw_weights = ef_ema_semi.max_sharpe(risk_free_rate = rfr)
semi_ema_ms_portfolio = pd.Series(semi_ema_ms_raw_weights)

print('********************** Approach 2: Semicovariance + EMA **********************')

# Minimum volatility portfolio performance
semi_ema_ms_portfolio_performance = ef_ema_semi.portfolio_performance(verbose = True, risk_free_rate = rfr)

print(semi_ema_ms_portfolio)
```
